```python
import pandas as pd
import numpy as np
```

```python
train_inter = pd.read_csv('../dataset/raw/browsing_train.csv')
train_search = pd.read_csv('../dataset/raw/search_train.csv')
```

```python
test_inter = pd.read_csv('../dataset/raw/browsing_test.csv')
test_search = pd.read_csv('../dataset/raw/search_test.csv')
```

```python
train_inter.head()
```

```python
all_data = pd.concat([train_inter, test_inter], keys=['train', 'test'])
all_data.head()
```

```python
all_data.dropna(subset=['product_sku_hash'], inplace=True)
all_data = all_data[all_data['product_action'] != 'remove'][['session_id_hash', 'product_sku_hash', 'server_timestamp_epoch_ms']]
all_data.head()
```

```python
all_data.columns = ['user_id:token', 'item_id:token', 'timestamp:float']
```

```python
all_data['user_id:token'] = pd.factorize(all_data['user_id:token'])[0]
all_data['item_id:token'] = pd.factorize(all_data['item_id:token'])[0]
```

```python
all_data.head()
```

```python
all_data.reset_index(drop=True, inplace=True)
```

```python
all_data.to_csv('../dataset/processed/sigir.inter', index=False, sep='\t')
```
