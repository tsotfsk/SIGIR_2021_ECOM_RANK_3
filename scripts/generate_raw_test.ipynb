```python
import pandas as pd
import numpy as np
```

```python
train_inter = pd.read_csv('../dataset/raw/browsing_train.csv')
test_merged = pd.read_csv('../dataset/raw/browsing_test.csv')
```

```python
train_search = pd.read_csv('../dataset/raw/sku_to_content.csv')
```

# 分离session 和 search

```python
inter_cols = train_inter.columns
search_cols = test_merged.columns
```

```python
search_mask = test_merged['is_search'] == True
test_search = test_merged[search_mask][search_cols]
test_search.head()
```

```python
inter_mask = test_merged['is_search'] == False
test_inter = test_merged[inter_mask][inter_cols]
test_inter.head()
```

```python
assert test_search.shape[0] + test_inter.shape[0] == test_merged.shape[0]
```

```python
test_inter.to_csv('../dataset/raw/browsing_test.csv', index=False)
test_search.to_csv('../dataset/raw/search_test.csv', index=False)
```
